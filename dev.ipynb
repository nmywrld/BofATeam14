```python
import csv

from orderNode import orderNode
```

```python
def csv_to_ArrayDict(filename):
    with open(filename, mode='r') as infile:
        reader = csv.DictReader(infile)
        data = [row for row in reader]
    return data

def csv_to_dict(filename):
    with open(filename, mode='r') as infile:
        reader = csv.DictReader(infile)
        data = {row[next(iter(reader.fieldnames))]: {key: value for key, value in row.items() if key != next(iter(reader.fieldnames))} for row in reader}
    return data
```

```python

client_data = csv_to_dict('./test-set/input_clients.csv')
print(client_data)
```

```python

inst_data = csv_to_dict('./test-set/input_instruments.csv')
print(inst_data)
```

```python

order_data = csv_to_ArrayDict('./test-set/input_orders.csv')
print(order_data)
```

```python
# node = orderNode(time, order_id, instrument, quantity, client_id, price, buy_or_sell, client_rating, client_position_check)

# loop through order_data and create orderNodes and put it inside a variable called orderBook

buyHeap = []
sellHeap = []
buyHeapMarket = []
sellHeapMarket = []

for order in order_data:

    if order['Time'] == "09:30:00":
        break

    if order['Side'] == "Sell":
        currObj = orderNode(order['Time'], order['OrderID'], order['Instrument'], order['Quantity'], order['Client'], order['Price'], order['Side'], client_data[order['Client']]["Rating"], client_data[order['Client']]["PositionCheck"])
        
        if order['Price'] == "Market":
            sellHeapMarket.append(currObj)
        else:
            sellHeap.append(currObj)

    else:
        currObj = orderNode(order['Time'], order['OrderID'], order['Instrument'], order['Quantity'], order['Client'], order['Price'], order['Side'], client_data[order['Client']]["Rating"], client_data[order['Client']]["PositionCheck"])
        
        if order['Price'] == "Market":
            buyHeapMarket.append(currObj)
        else:
            buyHeap.append(currObj)



```

```python
for order in buyHeap:
    print(order.time, order.order_id, order.instrument, order.quantity, order.client_id, order.price, order.buy_or_sell, order.client_rating, order.client_position_check)
    print()
```

```python
for order in buyHeapMarket:
    print(order.time, order.order_id, order.instrument, order.quantity, order.client_id, order.price, order.buy_or_sell, order.client_rating, order.client_position_check)
    print()
```
